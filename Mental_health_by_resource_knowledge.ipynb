Graphic displaying mental health seeking by knowledge of mental health resources. 

```python
# import necessary packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
df= pd.read_csv('Full_data.csv')
```

```python
g = sns.factorplot('treatment_num', data=df, hue='benefit_sum', palette='coolwarm', kind='count')
g.set_xticklabels(labels = ['No treatment sought', 'Treatment sought'])
g.set(xlabel='Mental health treatment', ylabel='Count')
g._legend.set_title('Sum of Benefits')
```
